```python
"""This code plot fire data  using basemap"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import pandas as pd
import io
import matplotlib.gridspec as gridspec


#=== Read data====
win = pd.read_csv('winter.csv')
spr = pd.read_csv('spring.csv')
mon = pd.read_csv('monsoon.csv')
aut = pd.read_csv('autumn.csv')

lat = win['latitude'].values
lon = win['longitude'].values

lat1 = spr['latitude'].values
lon1 = spr['longitude'].values

lat2 = mon['latitude'].values
lon2 = mon['longitude'].values

lat3 = aut['latitude'].values
lon3 = aut['longitude'].values


#==== plot============
fig, axs = plt.subplots(2, 2,figsize=(6,4),dpi=200)

#==== Winter
gridspec.GridSpec(2,2)
plt.subplot2grid((2,2), (0,0))
m = Basemap(projection='lcc',width=12000000,height=9000000,area_thresh=1000,\
            lat_1=25,lat_2=25,lat_0=25,lon_0=90,llcrnrlon=45,llcrnrlat=0,\
             urcrnrlon=139,urcrnrlat=55,resolution='h')

m.shadedrelief()
#m.etopo()
m.drawparallels(np.arange(10, 60, 15), linewidth=0.4, dashes=[4, 1], labels=[1,0,0,0], fontsize=8, color='black')
m.drawmeridians(np.arange(45, 125,15),linewidth=0.4, dashes=[4, 1],fontsize=10, color='black')
m.drawcountries()
m.drawcoastlines(linewidth=0.3)
m.fillcontinents(color = 'dimgrey') #,lake_color='#46bcec')

lons, lats = m(lon, lat)
m.scatter(lons, lats, marker = 'o', s=0.005,alpha=0.5,color='pink', zorder=5)
plt.title('Winter',fontsize=8)
#==== spring
plt.subplot2grid((2,2), (0,1))
m = Basemap(projection='lcc',width=12000000,height=9000000,area_thresh=1000,\
            lat_1=25,lat_2=25,lat_0=25,lon_0=90,llcrnrlon=45,llcrnrlat=0,\
             urcrnrlon=139,urcrnrlat=55,resolution='h')

m.shadedrelief()
#m.etopo()
m.drawparallels(np.arange(10, 60, 15), linewidth=0.4, dashes=[4, 1], fontsize=8, color='black')
m.drawmeridians(np.arange(45, 125,15),linewidth=0.4, dashes=[4, 1],fontsize=8, color='black')
m.drawcountries()
m.drawcoastlines(linewidth=0.3)
m.fillcontinents(color = 'dimgrey') #,lake_color='#46bcec')

lons, lats = m(lon1, lat1)
m.scatter(lons, lats, marker = 'o', s=0.005,alpha=0.5,color='pink', zorder=5)
plt.title('Spring',fontsize=8)

#==== summer
plt.subplot2grid((2,2), (1,0))
m = Basemap(projection='lcc',width=12000000,height=9000000,area_thresh=1000,\
            lat_1=25,lat_2=25,lat_0=25,lon_0=90,llcrnrlon=45,llcrnrlat=0,\
             urcrnrlon=139,urcrnrlat=55,resolution='h')

m.shadedrelief()
#m.etopo()
m.drawparallels(np.arange(10, 60, 15), linewidth=0.4, dashes=[4, 1], labels=[1,0,0,0], fontsize=8, color='black')
m.drawmeridians(np.arange(45, 125,15),linewidth=0.4, dashes=[4, 1],  labels=[0,0,0,1],fontsize=8, color='black')
m.drawcountries()
m.drawcoastlines(linewidth=0.3)
m.fillcontinents(color = 'dimgrey') #,lake_color='#46bcec')

lons, lats = m(lon2, lat2)
m.scatter(lons, lats, marker = 'o', s=0.005,alpha=0.5,color='pink', zorder=5)
plt.title('Monsoon',fontsize=8)

#==== Autumn
plt.subplot2grid((2,2), (1,1))
m = Basemap(projection='lcc',width=12000000,height=9000000,area_thresh=1000,\
            lat_1=25,lat_2=25,lat_0=25,lon_0=90,llcrnrlon=45,llcrnrlat=0,\
             urcrnrlon=139,urcrnrlat=55,resolution='h')

m.shadedrelief()
#m.etopo()
m.drawparallels(np.arange(10, 60, 15), linewidth=0.4, dashes=[4, 1], fontsize=10, color='black')
m.drawmeridians(np.arange(45, 125,15),linewidth=0.4, dashes=[4, 1],labels=[0,0,0,1],fontsize=8, color='black')
m.drawcountries()
m.drawcoastlines(linewidth=0.3)
m.fillcontinents(color = 'dimgrey') #,lake_color='#46bcec')

lons, lats = m(lon3, lat3)
m.scatter(lons, lats, marker = 'o', s=0.005,alpha=0.5,color='pink', zorder=5)
plt.title('Autumn',fontsize=8)



fig.subplots_adjust(top=0.95,
                        bottom=0.055,
                        left=0.055,
                        right=0.985,
                        hspace=0.115,
                        wspace=0.06)



#plt.savefig("/mnt/e/LaTeX-PPT/fire.png",dpi=200)
plt.show()
```
